```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
work = np.random.normal(50, 12, (200, 2)) % 100
home = np.random.normal(50, 30, (1000, 2)) % 100
stations = np.random.uniform(0, 100, (40, 2))
plt.plot(work[:, 0], work[:, 1], "b.", label = "Work Locations")
plt.plot(home[:, 0], home[:, 1], "y.", label = "Home Locations")
plt.plot(stations[:, 0], stations[:, 1], "k.", label = "Charging Stations")
plt.legend(loc=(1.04, .785))
plt.xlim(0, 100)
plt.ylim(0, 100)
plt.title("Downtown Model Initialization")
```
